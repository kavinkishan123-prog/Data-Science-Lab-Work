```python
# Problem 1: Confusion Matrix Metrics
TP = 50
FP = 8
TN = 32
FN = 10

total = TP + FP + TN + FN
accuracy = (TP + TN) / total
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1_score = 2 * precision * recall / (precision + recall)

print("Total Predictions:", total)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1_score)
```

```python
# Problem 2: R Squared
SSR = 420
SSE = 180

SST = SSR + SSE
R_squared = SSR / SST

print("Total Sum of Squares:", SST)
print("R Squared:", R_squared)
print("Explains more than 70% variance:", R_squared > 0.7)
```

```python
# Problem 3: Mean Squared Error
e1 = 4
e2 = 9
e3 = 16
e4 = 25
e5 = 36

MSE = (e1 + e2 + e3 + e4 + e5) / 5
print("Mean Squared Error:", MSE)
```

```python
# Problem 4: Cook's Distance
ei_squared = 9
MSE = 3
p = 3
hii = 0.2
n = 25

cooks_distance = (ei_squared / (p * MSE)) * (hii / ((1 - hii) ** 2))
threshold = 4 / n

print("Cook's Distance:", cooks_distance)
print("Threshold:", threshold)
print("Influential Observation:", cooks_distance > threshold)
```
