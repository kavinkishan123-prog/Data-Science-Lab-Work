## 1) Data Access
Load the datasets into 3 dataframes (CSV, TXT, XLSX) and compare them.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load CSV file
df_csv = pd.read_csv("Facebook Friends.csv")
df_csv.head()
```

```python
df_txt = pd.read_csv("Facebook Friends.txt", delimiter="\t", encoding="utf-16")
df_txt.head()
```

```python
# Load Excel file
df_xlsx = pd.read_excel("Facebook Friends.xlsx")
df_xlsx.head()
```

```python
# Compare shapes
print("CSV Shape:", df_csv.shape)
print("TXT Shape:", df_txt.shape)
print("XLSX Shape:", df_xlsx.shape)
```

```python
# Compare column names
print("CSV Columns:\n", df_csv.columns)
print("\nTXT Columns:\n", df_txt.columns)
print("\nXLSX Columns:\n", df_xlsx.columns)
```

### Conclusion (Data Source Differences)
The data source format **can make a difference**.

- CSV and TXT may require correct delimiters (comma/tab). If not specified correctly, columns can shift.
- Excel files usually preserve the table structure better.
- If the files are properly formatted, they should contain the same data.

## 3) Descriptive Statistics
Using `describe()` to compute summary statistics.


```python
# Basic descriptive statistics for numerical columns
df.describe()
```

```python
# Descriptive statistics for all columns
df.describe(include="all")
```

### Explanation of describe()
`describe()` provides:
- **count**: non-missing values
- **mean**: average
- **std**: standard deviation
- **min**: minimum
- **25%, 50%, 75%**: quartiles (median at 50%)
- **max**: maximum

This helps us understand the distribution and spread of the dataset.

## 4) Data Visualization
Histograms and frequency plots help identify distribution, outliers, and patterns.


```python
# Histograms for all numeric columns
df.hist(figsize=(18,12), bins=20)
plt.tight_layout()
plt.show()
```

```python
# Frequency plot of Age
df["Age"].value_counts().sort_index().plot(kind="bar", figsize=(14,5))
plt.title("Frequency Plot of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Histogram of Friends count
df["Friends"].plot(kind="hist", bins=30, figsize=(10,5))
plt.title("Histogram of Friends Count")
plt.xlabel("Friends")
plt.ylabel("Frequency")
plt.show()
```

```python
# Boxplot to detect outliers
df.boxplot(figsize=(18,6))
plt.xticks(rotation=90)
plt.title("Boxplot of Dataset Variables")
plt.show()
```

```python
# Correlation heatmap
import seaborn as sns

plt.figure(figsize=(14,10))
sns.heatmap(df.corr(numeric_only=True), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

### Visualization Insights
- Histograms show whether variables are skewed or normally distributed.
- Frequency plots show which values occur most often.
- Boxplots reveal outliers and extreme values.
- Correlation heatmap shows relationships between numeric variables.

Visualization makes it easier to understand the dataset compared to only reading tables.
